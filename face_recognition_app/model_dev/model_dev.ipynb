```python
import numpy as np
from keras_facenet import FaceNet
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
import os
```

```python
embedder = FaceNet()
```

```python
def directory_face_detection(root_directory):
    face_data = {}
    for root, _, files in os.walk(root_directory):
        for file in files:
            image_path = os.path.join(root, file)
            print(image_path)
            detection = embedder.extract(image_path)
            label = os.path.basename(root)
            # Add to our data structure
            if label not in face_data:
                face_data[label] = []
            face_data[label].append(detection)  
    return face_data
```

```python
student_image_face_data = directory_face_detection("../../data/images")
```

```python
import joblib
```

```python
joblib.dump(student_image_face_data, "student_image_face_data.pkl")
```

```python
len(student_image_face_data.keys())
```

```python
for key in student_image_face_data.keys():
    print(key)
```

```python
embeddings = []
labels = []
for key in student_image_face_data.keys():
    for image in student_image_face_data[key]:
        if len(image) != 0:
            embeddings.append(image[0]["embedding"])
            labels.append(key)
```

```python
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)
```

```python
encoded_labels
```

```python
joblib.dump(label_encoder, "label_encoder.pkl")
```

```python
classifier = SVC(kernel="linear", probability=True)
classifier.fit(embeddings, encoded_labels)
```

```python
joblib.dump(classifier, "face_recog_model.pkl")
```
